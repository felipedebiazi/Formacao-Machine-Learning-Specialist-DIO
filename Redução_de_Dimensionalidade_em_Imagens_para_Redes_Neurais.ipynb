```python
from PIL import Image, ImageFilter
import cv2
from skimage import io
from pylab import *
```

```python
rosto = '/content/gdrive/MyDrive/Dataset-DIO/rosto.jpg'
```

```python
image = Image.open(rosto)
image.resize((200,300))
```

```python
def greyscale(colored):
    w, h = colored.size
    img = Image.new("RGB", (w,h))

    for x in range(w):
        for y in range (h):
            pxl = colored.getpixel((x,y))
            # pesos das coordenadas RGB
            lum = int (0.3*pxl[0] + 0.59*pxl[1] + 0.11*pxl[2])
            img.putpixel((x,y), (lum, lum, lum))
    return img            
```

```python
pb_rosto = greyscale(image)
pb_rosto.resize((200,300))
```

```python
imagem = cv2.imread(rosto)
gray= cv2.cvtColor(imagem,cv2.COLOR_RGB2GRAY)
thresh = 100
binaryimage = cv2.threshold(gray,thresh,255,cv2.THRESH_BINARY)[1]
io.imshow(binaryimage)
```
